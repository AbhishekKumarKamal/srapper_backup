```python
def get_image_through_wiki_url(url):
    try:
        driver.get(url)
        time.sleep(1)
        WebDriverWait(driver,1)
        try:
            image = driver.find_element_by_xpath("//*[@id='mw-content-text']/div/table[2]/tbody/tr[2]/td/a/img").get_attribute('src')
            return image
        except:
            try:
                html = driver.execute_script("return document.documentElement.outerHTML;")
                soup = BeautifulSoup(html, 'lxml')
                try:
                    image = soup.find("table", attrs = {"class":"infobox vcard plainlist"}).find("tbody").find("td",attrs = {"colspan": 2}).find("a",attrs = {"class":"image"}).find("img").get('src')
                    #print("image ",image)
                    return "https:"+image
                except:
                    try:
                        image = soup.find("table", class_ = "infobox biography vcard").find("tbody").find("td",attrs = {"colspan": 2}).find("a",attrs = {"class":"image"}).find("img").get('src')
                        return "https:"+image
                    except:
                        try:
                            image = soup.find("div",class_ = "thumbinner").find("img",class_ = "thumbimage").get("src")
                            return "https:"+image
                        except Exception as e:
                            #print(e)
                            try:
                                img = soup.findAll("table", class_ = "infobox vcard")
                            except:
                                img = soup.findAll("table", class_ = "infobox biography vcard")
                    image = "https:"+img[0].find("img").get('src')
                    return image
            except:
                print("level 3")
                return []
    except:
        return []
```
